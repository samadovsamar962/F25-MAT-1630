```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# -----------------------------
# 1. Dataset (from your slide)
# -----------------------------
X = np.array([
    [1,2],
    [3,3],
    [3,5],
    [5,4],
    [5,6],
    [6,5],
    [8,7],
    [9,8]
])

# -----------------------------
# 2. Compute Covariance Matrix
# -----------------------------
cov_matrix = np.cov(X.T)
print("Covariance Matrix:")
print(cov_matrix)

# -----------------------------
# 3. Compute eigenvalues/eigenvectors manually
# -----------------------------
eigvals, eigvecs = np.linalg.eig(cov_matrix)
print("\nEigenvalues:")
print(eigvals)

print("\nEigenvectors:")
print(eigvecs)

# -----------------------------
# 4. PCA using scikit-learn
# -----------------------------
pca = PCA()
pca.fit(X)

print("\nPCA Eigenvalues (variance explained):")
print(pca.explained_variance_)

print("\nPCA Eigenvectors (principal components):")
print(pca.components_)

# -----------------------------
# 5. Plot data + PCA directions
# -----------------------------
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], label="Data Points")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("PCA Visualization")

# Center point (mean of data)
origin = np.mean(X, axis=0)

# Plot PCA vectors
for length, vector in zip(pca.explained_variance_, pca.components_):
    plt.arrow(origin[0], origin[1],
              vector[0] * length * 2,
              vector[1] * length * 2,
              head_width=0.2,
              color='red')

plt.grid(True)
plt.show()
```
